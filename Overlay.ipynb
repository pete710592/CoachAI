```python
import os
import glob
import cv2
import numpy as np
import pandas as pd
from utils.functions import checkDirExist

player_dir = os.path.join(os.getcwd(), 'dataset/evaluation/1_player/')
player_list = glob.glob(player_dir + "*.csv")

match_dir = os.path.join(os.getcwd(), 'dataset/position_5/trajectory-100/')
match_list = glob.glob(match_dir + "*.csv")

machine_dir = os.path.join(os.getcwd(), 'dataset/evaluation/2_machine/')
machine_list = glob.glob(machine_dir + "*.csv")

df = pd.read_csv('predictions/evaluation_result.csv')
df
```

```python
def plot(img, df, point_color):
    for index in range(len(df['X'])):
        y, x = df.loc[index, 'X'], df.loc[index, 'Y']
        
        # draw point as circle
        for i in range(-1, 2):
            for j in range(-1, 2):
                try:
                    img[x+i, y+j] = point_color
                except IndexError as e:
                    pass
    return img
    

def main():
    for idx in df.index:
        # Read csv file
        player = pd.read_csv(player_dir+df.loc[idx, 'player'])
        match = pd.read_csv(match_dir+df.loc[idx, 'match'])
        machine = pd.read_csv(machine_dir+df.loc[idx, 'model predict'])
        
        # Plot
        img = np.ones((1080, 1920, 3), np.uint8)*255
        plot(img, player, [255, 0, 0])
        plot(img, match, [0, 255, 0])
        plot(img, machine, [0, 0, 255])
        
        # Save as image
        cv2.imwrite('predictions/evaluation_%02d.jpg'%(idx+1), img)

if __name__ == "__main__":
    main()
```
